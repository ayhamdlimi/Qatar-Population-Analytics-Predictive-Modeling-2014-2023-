## How is the total population geographically distributed by municipality?

```python
df['Total'] = pd.to_numeric(df['Total'], errors='coerce').fillna(0)
```

```python
municipality_sum = df.groupby('Municipality')['Total'].sum().sort_values(ascending=False)
municipality_sum.plot(kind='bar', color='skyblue')
plt.title('Total Population by Municipality')
plt.ylabel('Population Count')
plt.show()
```

## How has the total population of each municipality evolved year-by-year from 2014 to 2023?

```python
yearly_growth = df.groupby(['Year', 'Municipality'])['Total'].sum().unstack()
yearly_growth.plot(figsize=(12, 6))
plt.title('Population Growth (2014 - 2023)')
plt.show()
```

## What was the population growth rate for each municipality between 2014 and 2023?

```python
yearly_diff = df.pivot_table(index='Year', columns='Municipality', values='Total', aggfunc='sum').diff()
max_growth_year = yearly_diff['Al Khor'].idxmax()
print(f"The largest population increase in Doha was in the year: {max_growth_year}")
```

```python
age_dist = df.groupby('Age_Group')['Total'].sum().sort_values(ascending=False)
age_dist.plot(kind='pie', autopct='%1.1f%%', figsize=(8,8))
plt.title('Population Distribution by Age Group')
```

## What is the percentage distribution of the total population across all municipalities in Qatar for the year 2023?

```python
total_pop_2023 = df[df['Year'] == 2023]['Total'].sum()
municipality_sums_2023 = df[df['Year'] == 2023].groupby('Municipality')['Total'].sum()
municipality_shares = (municipality_sums_2023 / total_pop_2023) * 100
(" Population Percentage Share by Municipality (2023)")
(municipality_shares.sort_values(ascending=False).map('{:.2f}%'.format))
```

## The projected population of Doha in 2025 is ... ML

```python
doha_data = df[df['Municipality'] == 'Doha'].groupby('Year')['Total'].sum().reset_index()
X = doha_data['Year'].values.reshape(-1, 1)
y = doha_data['Total'].values
model = LinearRegression()
model.fit(X, y)
future_year = np.array([[2025]])
prediction = model.predict(future_year)
(f"The projected population of Doha in 2025 is: {int(prediction[0])}")
```

## Predicting Qatar's population in 2026: ////// The current population is 3,370,611

```python
yearly_totals = df.groupby('Year')['Total'].sum().reset_index()
X = yearly_totals['Year'].values.reshape(-1, 1)
y = yearly_totals['Total'].values
model = LinearRegression()
model.fit(X, y)
future_year = [[2026]]
prediction_2026 = model.predict(future_year)
print(f"Predicted Population of Qatar in 2026: {int(prediction_2026[0]):,}")
score = model.score(X, y)
print(f"Model Accuracy (R² Score): {score:.4f}")
```

## Population Prediction per Age Group (2026)


```python
target_year = 2026
prediction_results = []
age_groups = df['Age_Group'].unique()
for age in age_groups:
    age_data = df[df['Age_Group'] == age].groupby('Year')['Total'].sum().reset_index()
    if len(age_data) > 1:
        X = age_data['Year'].values.reshape(-1, 1)
        y = age_data['Total'].values      
        model = LinearRegression()
        model.fit(X, y)     
        pred_value = model.predict([[target_year]])[0]
        prediction_results.append({
            'Age_Group': age,
            f'Predicted_Population_{target_year}': int(max(0, pred_value))
        })
df_age_predictions = pd.DataFrame(prediction_results)
print(f" Population Forecast per Age Group for {target_year} ")
(df_age_predictions.sort_values(by=f'Predicted_Population_{target_year}', ascending=False))
```
